```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
life = pd.read_csv("archive (1).zip")
```

```python
sns.barplot(x="Year",y="Alcohol",hue="Status",data=life)
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
sns.barplot(x="Year",y="Adult Mortality",hue="Status",data=life,color='aqua')
plt.xticks(rotation=45,ha="right")
plt.show()
```

```python
sns.distplot(life['percentage expenditure'])
plt.show()
```

```python
import numpy as np
sns.jointplot(life['GDP'])
plt.grid("TRUE")
plt.show()
```

```python
sns.histplot(x="Year",y="GDP",hue="Status",data=life)
plt.show()
```

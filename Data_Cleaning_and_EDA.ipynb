```python
import plotly.graph_objects as go

# Sample data (replace with real movie data)
movies = [
    {"title": "Movie A", "genres": ["Action", "Adventure", "Sci-Fi"]},
    {"title": "Movie B", "genres": ["Drama", "Romance"]},
    {"title": "Movie C", "genres": ["Action", "Sci-Fi", "Thriller"]},
    # Add more movies...
]

# Create nodes and connections
nodes = []
edges = []

for i, movie in enumerate(movies):
    nodes.append(dict(id=i, label=movie["title"], group=movie["genres"]))

    for j in range(i + 1, len(movies)):
        common_genres = set(movie["genres"]).intersection(movies[j]["genres"])
        if common_genres:
            edges.append(dict(source=i, target=j, value=len(common_genres)))

# Create the 3D plot
fig = go.Figure(data=[
    go.Scatter3d(
        x=[],
        y=[],
        z=[],
        mode='markers',
        marker=dict(size=8, color=[node["group"] for node in nodes], opacity=0.8),
        text=[node["label"] for node in nodes],
    ),
    go.Scatter3d(
        x=[edge["source"] for edge in edges],
        y=[edge["target"] for edge in edges],
        z=[edge["value"] for edge in edges],
        mode='lines',
        line=dict(color='rgba(125, 125, 125, 0.2)', width=1),
    ),
])

fig.update_layout(scene=dict(zaxis=dict(range=[0, 10])))

# Display the plot
fig.show()
```
